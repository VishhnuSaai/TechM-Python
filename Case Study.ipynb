```python
from pprint import * 
from collections import defaultdict
class Account:
    database= defaultdict(dict)

    def __init__(self,name,address,mobileno,aadhar,balance):
        self.id=mobileno
        self.name=name
        self.address=address
        self.mobileno=mobileno
        self.aadhar=aadhar
        self.balance=balance

        data={}

        for i in ['name','address','mobileno','aadhar','balance']:
            data[i]=eval(f'self.{i}')

        self.database[self.id]=data


        
def create_account(name,address,mobileno,aadhar,balance):
    return Account(name,address,mobileno,aadhar,balance)


while True:
    choice=int(input(
        """
        1.Create Account
        2.View Account Details
        3.Withdraw
        4.Deposit
        5.Fund Transfer
        6.Print Transactions
        7.Exit
        """
        ))

    if choice==1:
        name=input("Enter Name:")
        address=input("Enter Address")
        mobileno=input("Enter mobile number:")
        aadhar=input("Enter aadhaar number:")
        balance=100
        create_account(name,address,mobileno,aadhar,balance)

    elif choice==2:
        mob=input("Enter Mobile number:")
        print("Your account details are:")
        for i in ['name','address','mobileno','aadhar','balance']:
            print("{}:".format(i),Account.database[mob][i])

    elif choice==3:
        mob=input("Enter your id(mobile number):")
        amount=int(input("Enter amount to withdraw:"))
        Account.database[mob]['balance']-=amount
        print("Your current balance is:",Account.database[mob]['balance'])

    elif choice==4:
        mob=input("Enter your id(mobile number):")
        amount=int(input("Enter amount to deposit"))
        Account.database[mob]['balance']+=amount
        print("Your current balance is:",Account.database[mob]['balance'])

    elif choice==5:
        mob=input("Enter your id(mobile number):")
        receiver=input("Enter id of receiver(mobile no of receiver): ")
        if receiver not in Account.database:
            print("Invalid receiver credentials")
        else:
            amount=int(input("Enter amount to send:"))
            Account.database[mob]['balance']-=amount
            print("Amount succesfully sent")
            print("Your current balance is:",Account.database[mob]['balance'])
            Account.database[receiver]['balance']+=amount
    elif choice==6:
        mob=input("Enter your id(mobile number) for transactions:")


    elif choice==7:
        break












    




    
    
```
